```python
#descobre se o campeão foi o melhor ataque e ou a melhor defesa


def pega_linha_campeão(html_yy_yy, nome_do_campeonato):   # transformando em uma função que funcione com qualquer campeonato de qualquer ano
#recebe uma variavel com a tabela de uma temporada.
  melhor_ataque = True #inicia as variaveis como verdadeiras até provado o contrario
  melhor_defesa = True

  for i in range(len(html_yy_yy)):# vai de linha em linha checando as condições

    if melhor_ataque: #vai executar de qualquer jeito
      if html_yy_yy.iloc[0,6] < html_yy_yy.iloc[i,6]: # se a os gols pro do campeão for menor que qualquer outro time, melhor ataque recebe false
        melhor_ataque = False

    if melhor_defesa:
      if html_yy_yy.iloc[0,7] > html_yy_yy.iloc[i,7]: # se a os gols contra do campeão for maior que qualquer outro time, melhor defesa recebe false
        melhor_defesa = False

  if (melhor_ataque == False) and (melhor_defesa == True):# se melhor ataque é falso e melhor defesa verdadeiro, cria uma coluna "melhor" que recebe "defesa"
    print('melhor defesa')
    linha = html_yy_yy.iloc[[0]]
    linha["Melhor"] = "Defesa"

  if (melhor_ataque == True) and (melhor_defesa == False):#viceversa
    print('melhor ataque')
    linha = html_yy_yy.iloc[[0]]
    linha["Melhor"] = "Ataque"

  if (melhor_ataque == False) and (melhor_defesa == False):# se nao for o melhor em nada "melhor" recebe 0
    print("nao é o melhor em nada")
    linha = html_yy_yy.iloc[[0]]
    linha["Melhor"] = 0

  if (melhor_ataque == True) and (melhor_defesa == True):# se for o melhor nos dois "melhor" recebe "Ataque e Defesa"
    print('melhor ataque e melhor defesa')
    linha = html_yy_yy.iloc[[0]]
    linha["Melhor"] = "Ataque e Defesa"

  linha["Campeonato"] = str(nome_do_campeonato)
  return linha #retorna a linha do campeao

```

**CRIANDO A BASE DE DADOS PARA O POWERBI/DASHBOARD**

```python
Base_de_dados_PBI_lista = list(map(lambda x: pega_linha_campeão(x, "Premier League ENG"), Campeonato)) # vai pegando de linha em linha o campeao da temporada,
#já adcionando o melhor atributo e o nome do campeonato
type(Base_de_dados_PBI_lista)
Brasileiro_B_PBI = pd.concat(Base_de_dados_PBI_lista, ignore_index=1)#junta essa lista com os campeões em 1 data frame do campeonato em questao
```
